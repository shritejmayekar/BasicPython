```python
from matplotlib import pyplot as plt
import numpy as  np
plt.style.use("fivethirtyeight")
cs_plt_y = [10,40,60,65,90]
extc_plt_y = [20,10,75,45,30]
mech_plt_y = [30,20,10,10,32]

plt_x = ['Jan','Feb','Mar','Apr','May']
x_indexes = np.arange(len(plt_x))
width=0.25
fig, ax = plt.subplots()
rects1 = ax.bar(x_indexes-width/2,cs_plt_y,width=width,color='orange', label="CS")
rects2 = ax.bar(x_indexes+width/2 ,extc_plt_y,width=width,color='blue', label="EXTC")
rects3 = ax.bar(x_indexes+width*3/2 ,mech_plt_y,width=width,color='red', label="MECH")
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('No. of Registration')
ax.set_xlabel('Months')
ax.set_title('Students Registration By Month')
ax.set_xticks(x_indexes)
ax.set_xticklabels(plt_x)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()
plt.show()
```
