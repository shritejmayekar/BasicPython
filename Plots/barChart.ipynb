```python
from matplotlib import pyplot as plt
import numpy as  np
plt.style.use("fivethirtyeight")
cs_plt_y = [50,40,60,65,90]
extc_plt_y = [20,10,75,45,30]
mech_plt_y = [30,20,10,10,32]

plt_x = ['Jan','Feb','Mar','Apr','May']
x_indexes = np.arange(len(plt_x))
width=0.25
plt.bar(plt_x,cs_plt_y,width=width,color='orange', label="CS")
plt.bar(x_indexes-width ,extc_plt_y,width=width,color='blue', label="EXTC")
plt.bar(x_indexes+width ,mech_plt_y,width=width,color='red', label="MECH")

plt.title('Students Registration By Month')
plt.xlabel('Months')
plt.ylabel('No. of Registrations')

plt.legend()
plt.tight_layout()
plt.show()
```
