```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
slices = [50,40,60,65,90]
labels = ['Jan','Feb','Mar','Apr','May']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels=labels,shadow=True,startangle=90,
explode=explode,autopct='%1.1f%%',wedgeprops={'edgecolor':'blue'})


plt.title('Students Registration')
plt.tight_layout()
plt.show()
```
